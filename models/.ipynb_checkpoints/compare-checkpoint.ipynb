```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import xgboost as xgb
from scipy.stats import ttest_ind, mannwhitneyu
```

```python
# 1. Set project root and cleaned data directory
projekt_root = os.path.abspath(os.path.join(os.getcwd(), ".."))
cleaned_data_dir = os.path.join(projekt_root, "cleaned_data")

# 2. Define file paths for datasets
file_ai = os.path.join(cleaned_data_dir, "cleaned_global_ai_ml_data_salaries.csv")
file_ai_ml = os.path.join(cleaned_data_dir, "cleaned_global-salaries-in-ai-ml-data-science.csv")
file_traditional = os.path.join(cleaned_data_dir, "cleaned_Salary.csv")

# 3. Load datasets
df_ai = pd.read_csv(file_ai)
df_ai_ml = pd.read_csv(file_ai_ml)
df_traditional = pd.read_csv(file_traditional)
```

```python
# 4. Assign categories
df_ai["Category"] = "AI/ML"
df_ai_ml["Category"] = "AI/ML"
df_traditional["Category"] = "Traditional"

# 5. Merge AI datasets (Traditional salaries remain separate)
df_ai_combined = pd.concat([df_ai, df_ai_ml], ignore_index=True)

# 6. Function to detect column names dynamically
def find_column_by_keyword(df, keywords):
    for col in df.columns:
        if any(keyword.lower() in col.lower() for keyword in keywords):
            return col
    return None
```

```python
# 7. Identify Year and Salary columns
year_col = find_column_by_keyword(df_ai_combined, ["year", "work_year"])
salary_col_ai = find_column_by_keyword(df_ai_combined, ["salary", "annual salary", "base salary"])
salary_col_traditional = find_column_by_keyword(df_traditional, ["salary", "annual salary", "base salary"])

# 8. Process Year column in AI dataset
if year_col:
    df_ai_combined.rename(columns={year_col: "Year"}, inplace=True)
    df_ai_combined["Year"] = pd.to_numeric(df_ai_combined["Year"], errors="coerce")
    df_ai_combined.dropna(subset=["Year"], inplace=True)
    print(f"Year column found: {year_col}, renamed to 'Year'")
else:
    raise ValueError("No Year column found in AI dataset.")
```

```python
# 9. Process Salary column in AI dataset
if salary_col_ai:
    df_ai_combined.rename(columns={salary_col_ai: "Salary"}, inplace=True)
    df_ai_combined["Salary"] = pd.to_numeric(df_ai_combined["Salary"], errors="coerce")
    df_ai_combined.replace([np.inf, -np.inf], np.nan, inplace=True)  # FIX: Convert infinite values
    df_ai_combined.dropna(subset=["Salary"], inplace=True)
    print(f"Salary column found in AI dataset: {salary_col_ai}, renamed to 'Salary'")
else:
    raise ValueError("No Salary column found in AI dataset.")
```

```python
# 10. Process Salary column in Traditional dataset
if salary_col_traditional:
    df_traditional.rename(columns={salary_col_traditional: "Salary"}, inplace=True)
    df_traditional["Salary"] = pd.to_numeric(df_traditional["Salary"], errors="coerce")
    df_traditional.replace([np.inf, -np.inf], np.nan, inplace=True)  # FIX: Convert infinite values
    df_traditional.dropna(subset=["Salary"], inplace=True)
    print(f"Salary column found in Traditional dataset: {salary_col_traditional}, renamed to 'Salary'")
else:
    raise ValueError("No Salary column found in Traditional dataset.")
```

```python
# 11. Extract AI/ML salaries for 2024
df_ai_2024 = df_ai_combined[df_ai_combined["Year"] == 2024]
# Remove extremely high salaries to avoid skewing the visualization
df_ai_2024 = df_ai_2024[df_ai_2024["Salary"] < 500000]  # Remove salaries above $500k
df_traditional = df_traditional[df_traditional["Salary"] < 500000]
# 12. Direct Salary Comparison: AI/ML (2024) vs Traditional Salaries
mean_ai_2024 = df_ai_2024["Salary"].mean()
mean_traditional = df_traditional["Salary"].mean()

print(f"\nMean Salary (AI/ML 2024): ${mean_ai_2024:,.2f}")
print(f"Mean Salary (Traditional): ${mean_traditional:,.2f}")
```

```python
# Perform statistical tests
t_stat, p_value = ttest_ind(df_ai_2024["Salary"], df_traditional["Salary"], equal_var=False)
u_stat, p_u = mannwhitneyu(df_ai_2024["Salary"], df_traditional["Salary"])

print(f"T-test: t-statistic = {t_stat:.2f}, p-value = {p_value:.4f}")
print(f"Mann-Whitney U test: U-statistic = {u_stat:.2f}, p-value = {p_u:.4f}")
```

```python
if p_value < 0.05:
    print("AI salaries in 2024 are significantly higher than traditional industry salaries.")
else:
    print("No significant difference between AI and traditional industry salaries in 2024.")
```

```python
# 13. AI Salary Growth vs. Fixed Traditional Salary
df_ai_yearly = df_ai_combined.groupby("Year")["Salary"].mean().reset_index()
df_ai_yearly["Traditional Salary"] = mean_traditional  # Assume traditional salary is constant
```

```python
# Plot salary trends over time
plt.figure(figsize=(12, 6))
sns.lineplot(x="Year", y="Salary", data=df_ai_yearly, label="AI/ML", marker="o")
plt.axhline(y=mean_traditional, color="red", linestyle="--", label="Traditional (Constant)")
plt.xlabel("Year")
plt.ylabel("Average Salary ($)")
plt.title("AI/ML vs Traditional Salaries Over Time")
plt.legend()
plt.show()
```

```python
# 14. Salary Distribution Analysis (Boxplot & Violin Plot)
plt.figure(figsize=(10, 6))
sns.boxplot(x="Category", y="Salary", data=pd.concat([df_ai_2024, df_traditional]))
plt.xlabel("Industry Category")
plt.ylabel("Annual Salary ($)")
plt.title("Salary Distribution: AI/ML vs Traditional Industries")
plt.show()
plt.figure(figsize=(10, 6))
sns.violinplot(x="Category", y="Salary", data=pd.concat([df_ai_2024, df_traditional]))
plt.xlabel("Industry Category")
plt.ylabel("Annual Salary ($)")
plt.title("Salary Density: AI/ML vs Traditional Industries")
plt.show()
```
